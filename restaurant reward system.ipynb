```python
df['OrderCompleted']=df['PercentageOfOrderCompleted']/100
```

```python
def tu(r):
    if r>=150:
        return 'Platinum'
    elif r>=100:
        return 'Gold'
    elif r>=75:
        return 'Silver'
    else:
        return 'Bronze'
```

```python
df['Badges']=df['monthly_score'].apply(tu)
```

```python
x=df[['revenue', 'OrderCompleted', 'RepeatCustomers','AvgTimeToPrepareOrder(mins)']]
y=df['Badges']
```
